There are not really useful strong Correlations from the Heatmap but we gonna try it with those

```python
# Creating a Scatterplot for the "num_vehicle_types_involved" and "hour" Solumns in seaborn

sns.lmplot(x = 'num_vehicle_types_involved', y = 'hour', data = df)
```

The scatterplot between hour and num_vehicle_types_involved reveals no overall trend, but a dense concentration of accidents around 13:00–14:00, with most involving only one or two vehicle types. Accidents involving three or more types of vehicles are relatively rare and do not show time-specific patterns.

```python
# Creating a Scatterplot for the "hour" and "accident_category" Columns in seaborn

sns.lmplot(x = 'hour', y = 'accident_category', data = df)
```

The scatterplot between hour and accident_category reveals a high concentration of accidents within one specific category, just below code 3 (minor injuries), across all hours of the day. This suggests that one type of accident is disproportionately more common, regardless of the time it occurs.

```python
# Creating a Scatterplot for the "distance_to_central_point_km" and "location_density_score" Columns in seaborn

sns.lmplot(x = 'distance_to_central_point_km', y = 'location_density_score', data = df)
```

Most values are crowded near the bottom left, there is a slight downward trend as distance increaes, density decreases. Suggests, that accidents happen more frequently in dense areas closer to the centre. Accident density is highest close to the city centre and declines with distance.

## 6. Categorial Plots

```python
# Create a Categorical Plot in seaborn

sns.set(style="ticks")
g = sns.catplot(x="weekday", y="num_vehicle_types_involved", hue="bike_and_car", data=df)
```

Accidents involving both bikes and cars occur on all weekdays but mostly involve just one vehicle type. There is no noticeable variation by weekday, suggesting that bike-and-car incidents are not strongly tied to specific days of the week.

```python
# Create a Categorical Plot in seaborn

sns.set(style="ticks")
g = sns.catplot(x="season", y="accident_category", hue="is_multiple_vehicle", data=df)
```

Accident severity categories are stable across seasons, with most accidents being minor (category 1). Multiple-vehicle accidents occur in all seasons but do not appear disproportionately in any specific severity class.
